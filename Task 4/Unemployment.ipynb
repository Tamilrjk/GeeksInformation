The Unemployment Rate Analysis project focuses on leveraging data science to analyze the unemployment rate, especially in light of the significant surge caused by the Covid-19 pandemic. By studying the unemployment rate trends and patterns, this project aims to gain valuable insights into the impact of the pandemic on the workforce and the economy. The goal is to provide a comprehensive analysis that aids policymakers and organizations in making informed decisions to address unemployment challenges effectively.

```python
# Print the columns in df1
print("Columns in df1:", df1.columns)

# Print the columns in df2
print("Columns in df2:", df2.columns)
```

```python
# Strip leading and trailing whitespaces from column names
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

# Now try the merge again
df = pd.merge(df1, df2, on=['Date', 'Region'], how='inner')
```

```python
# Rename columns
df = df.rename(columns={
    'Frequency_x': 'Frequency',
    'Estimated Unemployment Rate (%)_x': 'Unemployment Rate_x',
    'Estimated Employed_x': 'Employed_x',
    'Estimated Labour Participation Rate (%)_x': 'Labour Participation Rate_x',
    'Frequency_y': 'Frequency',
    'Estimated Unemployment Rate (%)_y': 'Unemployment Rate_y',
    'Estimated Employed_y': 'Employed_y',
    'Estimated Labour Participation Rate (%)_y': 'Labour Participation Rate_y'
})
```

## Time Series

```python
#Explore the Time Series
plt.figure(figsize=(12, 6))
plt.plot(df['Unemployment Rate_x'])
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
#Decompose Time Series
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(df['Unemployment Rate_x'], model='additive', period=15)
result.plot()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

def check_stationarity(timeseries):
    result = adfuller(timeseries, autolag='AIC')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:', result[4])

check_stationarity(df['Unemployment Rate_x'])
```

```python
#Autocorrelation and Partial Autocorrelation
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(df['Unemployment Rate_x'], lags=20)
plt.show()

plot_pacf(df['Unemployment Rate_x'], lags=20)
plt.show()
```

```python
#Build ARIMA Model
from statsmodels.tsa.arima.model import ARIMA

# Replace p, d, and q with the identified values
p = 1  # Order of AR component
d = 1  # Order of differencing
q = 1  # Order of MA component

model = ARIMA(df['Unemployment Rate_x'], order=(p, d, q))
results = model.fit()
print(results.summary())
```
